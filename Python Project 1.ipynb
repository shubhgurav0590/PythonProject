```python
# Gender
ax = sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.columns
```

```python
# Age group
ax = sns.countplot(x='Age Group',hue ="Gender",data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# total amount vs age group
sale_group = df.groupby(["Age Group"],as_index = False)['Amount'].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Age Group",y="Amount",data = sale_group)
```

```python
# total number of orders from 10 states
sale_order = df.groupby(["State"],as_index = False)['Orders'].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x="State",y="Orders",data = sale_order)
```

```python
# total amount/sales from  top 10 states9
sale_amount = df.groupby(["State"],as_index = False)["Amount"].sum().sort_values(by="Amount",ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x="State",y="Amount",data=sale_amount)
```

```python
# from both the graph we see that most of the orders are from UP,Maharashtra,Karnataka,Delhi& total amount spent  is also same for these city
```

```python
# Marital Status
ax = sns.countplot(x='Marital_Status',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
    
sale_status = df.groupby(["Marital_Status","Gender"],as_index = False)["Amount"].sum().sort_values(by="Amount",ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x="Marital_Status",y="Amount",data=sale_status,hue="Gender")
```

```python
# occupation
ax = sns.countplot(x="Occupation",data=df)
sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sale_occup = df.groupby(["Occupation"],as_index = False)["Amount"].sum().sort_values(by="Amount",ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x="Occupation",y="Amount",data=sale_occup)
```

```python
# From the above graph can see that most of buyers are working in IT Sector, HealthCare,Aviation
# Product Category
ax = sns.countplot(x="Product_Category",data=df)
sns.set(rc={'figure.figsize':(25,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sale_occup = df.groupby(["Product_Category"],as_index = False)["Amount"].sum().sort_values(by="Amount",ascending = False).head(10)
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(x="Product_Category",y="Amount",data=sale_occup)
```
